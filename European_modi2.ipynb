```python
def first_bet(true_num):
    '''
    用來計算一開始下注
    paras:
        p_card 傳入真數
    return
        bet    下注     type:int
    history:2022/10/9 done
    '''
    if true_num >=6:
        bet = 30
    elif true_num ==5:
        bet = 25
    elif true_num ==4:
        bet = 20
    elif true_num ==3:
        bet = 15
    elif true_num ==2:
        bet = 10
    elif true_num ==1:
        bet = 5
    else:
        bet = 1
    return bet  
```

```python
def game_start(true_num,weight):
    
    '''###game_start 這個Function主要用來遊戲開局，初始化每個人的牌庫狀態，並將狀態寫入simulation中
    Paras:
        true_num:該手結束的真數     type:int    ex:3  
        weight:該手結束的權重      type:int    ex:15
    Returns:
        None
    
    '''

    bet= first_bet(true_num)
    d_sts={'weight':0,'bet':bet,'card_big':0,'card_small':0,'card_total':0,'card_ACE':0,'card_point':0,'ACE_Point':0,'BJ':'N',
         'Bust':'N','double_available':'Y','Stand':'N','Split_available':'N','Split_count':0,'Split_ACE':'N','surrender':'N'}

    simulation['p1_sts']=simulation['p1'].apply(lambda row:bf_sts_cal([row][0],bet))
    simulation['p2_sts']=simulation['p2'].apply(lambda row:bf_sts_cal([row][0],bet))
    simulation['p3_sts']=simulation['p3'].apply(lambda row:bf_sts_cal([row][0],bet))
```

```python
def card_count(p_card):
    '''
    被呼叫來算點數、大小牌、用掉的牌數、以及有沒有ACE等等
    這裡產出的狀態會放在dataframe裡面，計算包含了BJ、兩張A可拆、一般牌可拆、一般牌可加倍
    paras:
        p_card 傳入要計算的牌組 type:List()，List裡面放card物件
    return
        card_total 總共的牌數 其實都是2
        card_big   總共的大牌
        card_small 總共的小牌
        card_ACE   總共的ACE數
        ACE_Point  依照有沒有ACE動態判斷點數。
    history:2022/10/9 done
    '''
    if p_card[1].point==1 or p_card[0].point==1:
        card_ACE = 1
        ACE_Point=p_card[1].point+p_card[0].point+10
    else:
        card_ACE = 1
        ACE_Point=0
    if p_card[0].weighted ==1:
        card_big = 0
        card_small = 1
    elif p_card[0].weighted == -1:
        card_big = 1
        card_small = 0
    else :
        card_big = 0
        card_small = 0
    if p_card[1].weighted ==1:
        card_big += 0
        card_small += 1
    elif p_card[1].weighted ==-1:
        card_big += 1
        card_small += 0
    else :
        card_big += 0
        card_small += 0    
    card_total=2
    
    return card_total,card_big,card_small,card_ACE,ACE_Point
```

```python
def bf_sts_cal(p_card,bet):
    '''
    用來計算一開始的兩張，總共會帶來的狀態，需要呼叫card_count，去算點數、大小牌、用掉的牌數、以及有沒有ACE等等
    這裡產出的狀態會放在dataframe裡面，計算包含了BJ、兩張A可拆、一般牌可拆、一般牌可加倍
    paras:
        p_card 傳入要計算的牌組 type:List()，List裡面放card物件
        bet    起始的注碼       type:int  
    return
        p_sts  該牌組的狀態     type:dict() 
    history:2022/10/9 done
    '''
    if (p_card[0].point+p_card[1].point==11) and (p_card[0].point or p_card[1].point==1):#只有兩張時加起來十一點且有一張為1，就是BJ
        p_sts={'weight':p_card[0].weighted+p_card[1].weighted,'bet':bet,'card_big':2,'card_small':0,'card_total':2,'card_ACE':1,'card_point':21,'ACE_Point':21,
                  'BJ':'Y','Bust':'N','double_available':'N','Stand':'N','Split_available':'N','Split_count':0,'Split_ACE':'N'
                  ,'surrender':'N'}
        return p_sts
    elif (p_card[0].point == p_card[1].point) and (p_card[0].point or p_card[1].point==1):#兩張A的拆牌
        p_sts={'weight':p_card[0].weighted+p_card[1].weighted,'bet':bet,'card_big':2,'card_small':0,'card_total':2,'card_ACE':2,'card_point':2,'ACE_Point':12,
                  'BJ':'N','Bust':'N','double_available':'N','Stand':'N','Split_available':'Y','Split_count':0,'Split_ACE':'Y'
                  ,'surrender':'N'}
        return p_sts
    elif (p_card[0].point == p_card[1].point) and (p_card[1].point!=1):#兩張其他的拆牌
        card_total,card_big,card_small,card_ACE,ACE_Point=card_count(p_card)
        p_sts={'weight':p_card[0].weighted+p_card[1].weighted,'bet':bet,'card_big':card_big,'card_small':card_small,
                'card_total':card_total,'card_ACE':0,'card_point':p_card[1].point*2,'ACE_Point':0,
                  'BJ':'N','Bust':'N','double_available':'N','Stand':'N','Split_available':'Y','Split_count':0,'Split_ACE':'N'
                  ,'surrender':'N'}
        return p_sts
    else:
        card_total,card_big,card_small,card_ACE,ACE_Point=card_count(p_card)
        p_sts={'weight':p_card[0].weighted+p_card[1].weighted,'bet':bet,'card_big':card_big,'card_small':card_small,
                'card_total':card_total,'card_ACE':0,'card_point':p_card[1].point+p_card[0].point,'ACE_Point':0,
                  'BJ':'N','Bust':'N','double_available':'Y','Stand':'N','Split_available':'N','Split_count':0,'Split_ACE':'N'
                  ,'surrender':'N'}
        return p_sts
```

```python
def card1_handle(tempcards,weight,p_sts,p_card):
  ''' ###card1_handle，準備每個玩家拿到第一張牌時變動的參數。
    Paras:
        tempcards 傳入目前抽出的那張牌 type:class 
        weight  目前的權重     type:int
        p_sts   當下處理牌組的狀況 type:dict
        p_card  第一張牌      type:list
    Returns:
        None
  '''
  p_sts['weight']=tempcards.weighted
  p_sts['card_point']=tempcards.point
  p_sts['ACE_Point']=tempcards.point
  if p_card[0].weighted ==1:
    p_sts['card_big'] = 0
    p_sts['card_small'] = 1
  elif p_card[0].weighted == -1:
    p_sts['card_big'] = 1
    p_sts['card_small'] = 0
  else :
    p_sts['card_big'] += 0
    p_sts['card_small'] += 0
  if p_card[0].ACE=='1':
    p_sts['card_ACE'] =1
  p_sts['card_total'] =len(p_card)
  return p_sts,p_card


```

```python
def card2_handle(tempcards,weight,p_sts,p_card):
  ''' ###card2_handle，準備每個玩家第二張牌開始時需要的參數跟card1差在索引直接選第二張，還有手牌狀態的調整。
    Paras:
        tempcards 傳入目前抽出的那張牌 type:class 
        weight  目前的權重     type:int
        p_sts   當下處理牌組的狀況 type:dict
        p_card  第一張牌      type:list
    Returns:
        None
  '''
  p_sts['weight']+=tempcards.weighted #權重
  p_sts['card_point']+=tempcards.point #點數計算
  split_cal(p_sts,p_card)
  if p_card[1].weighted ==1:
    p_sts['card_big'] += 0
    p_sts['card_small'] += 1
  elif p_card[1].weighted ==-1:
    p_sts['card_big'] += 1
    p_sts['card_small'] += 0
  else :
    p_sts['card_big'] += 0
    p_sts['card_small'] += 0
  if p_card[1].ACE=='1':
    p_sts['card_ACE'] +=1
    # print('Write card2 = ACE')
  p_sts['card_total'] =len(p_card)
  p_sts,p_card=Ace_cal(p_sts,p_card)#計算是否有ACE的點數，是否為BJ
  return p_sts,p_card
```

```python
def true_num_cal(weight):
  if int(weight/6)>=6:
    true_num =6
  elif int(weight/6)==5:
    true_num =5
  elif int(weight/6)==4:
    true_num =4
  elif int(weight/6)==3:
    true_num =3
  elif int(weight/6)==2:
    true_num =2
  elif int(weight/6)==1:
    true_num =1
  else:
    true_num =0
  return true_num
```

```python
def Ace_cal(p_sts,p_card):
  ''' ###Ace_cal，計算第二張牌以後如果有一張牌是ACE的話，用來比大小的ACE點數，以及是否有BJ。
    Paras:
        p_sts   當下處理牌組的狀況 type:dict
        p_card  玩家手牌      type:list
    Returns:
        None
  '''
  if p_sts['card_ACE']>=1:
    for i in range(len(p_card)):
      p_sts['card_point']+=p_card[i].point
    p_sts['ACE_Point']=p_sts['card_point']+10
  if p_sts['ACE_Point']==21 and len(p_card)==2 and p_sts['Split_count']==0:
    p_sts['BJ'] = 'Y'
  Bust(p_sts,p_card)
  return p_sts,p_card

    
  
```

```python
def split_cal(p_sts,p_card):
  ''' ###split_cal，計算是否後面是否可以分牌，有分ACE分牌跟一般分牌。
    Paras:
        p_sts   當下處理牌組的狀況 type:dict
        p_card  玩家手牌     type:list
    Returns:
        None
  '''
  if len(p_card)==2 and (p_card[0].point == p_card[1].point) and p_sts['Split_count']<2 and p_sts['Split_ACE']=='N':
    p_sts['Split_available'] = 'Y'
  elif len(p_card)==2 and (p_card[0].point == p_card[1].point) and p_sts['Split_count']<1 and p_card[1].ACE==1:
    p_sts['Split_available'] = 'Y'
  else:
    p_sts['Split_available'] = 'N'
```

```python
def check_slot(tempcard_slot,choose_slot):
  ''' ###check_slot，每發一張牌就需要重新確認牌夾還有沒有牌，不然都會超過index。
  如果沒牌了就要換一個新的牌夾，然後把dataframe的舊的牌夾row清空
    Paras:
        None
    Returns:
        tempcard_slot 新的牌夾號碼
  '''

  if len(tempcard_slot)>0:
    return tempcard_slot
  else:
    df.loc[choose_slot]=None
    choose_slot=ch_slot(df)
    tempcard_slot=list(filter(lambda item: item is not None, df.iloc[choose_slot].values.tolist()))
    return tempcard_slot
```

```python
def Bust(p_sts,p_card):
  ''' ###Bust，計算是否爆牌，若爆牌就結束了。
    Paras:
        p_sts   當下處理牌組的狀況 type:dict
        p_card  玩家手牌     type:list
    Returns:
        None
  '''
  if p_sts['ACE_Point']>21 and p_sts['card_point']>21:
    p_sts['Bust'] = 'Y'
  else:
    p_sts['Bust'] = 'N'
```

```python
def Double(p_sts,p_card,way,tempcard_slot,choose_slot,df):
  if p_sts['double_available']=='Y' :
    p_sts['bet']=p_sts['bet']*2
    tempcard_slot,choose_slot,p_sts,p_card=hit(p_sts,p_card,tempcard_slot,choose_slot,df)
  elif way=='Stand':
    p_sts['Stand'] ='Y'
    p_sts['double_available']='N'
  elif way == 'hit':
    choose_slot=ch_slot(df)
    tempcard_slot=list(filter(lambda item: item is not None, df.iloc[choose_slot].values.tolist()))
    tempcard_slot=check_slot(tempcard_slot,choose_slot)
    tempcard_slot,choose_slot,p_sts,p_card=hit(p_sts,p_card,tempcard_slot,choose_slot,df)
    
  return tempcard_slot,choose_slot,p_sts,p_card
```

```python
def hit(p_sts,p_card,tempcard_slot,choose_slot,df):
  choose_slot=ch_slot(df)
  tempcard_slot=list(filter(lambda item: item is not None, df.iloc[choose_slot].values.tolist()))
  tempcard_slot=check_slot(tempcard_slot,choose_slot)
  tempcard=tempcard_slot.pop() 
  p_card.append(tempcard)
  # print('again we are here')
  card2_handle(tempcard,p_sts['weight'],p_sts,p_card)
  del_card = df.loc[choose_slot].count().sum()-1
  df.loc[choose_slot][del_card]=None
  p_sts['double_available']='N'
  p_sts['Split_available']='N'
  Bust(p_sts,p_card)
  p_sts,p_card=Ace_cal(p_sts,p_card)
  # print('t_slot',tempcard_slot,'c_slot',choose_slot)
  # print(p_sts)
  return tempcard_slot,choose_slot,p_sts,p_card
```

```python
def surrender(p_sts,p_card,tempcard_slot,choose_slot,df):
  if len(p_card)==2:
    p_sts['surrender'] = 'Y'
    return tempcard_slot,choose_slot,p_sts,p_card
  else:
    choose_slot=ch_slot(df)
    tempcard_slot=list(filter(lambda item: item is not None, df.iloc[choose_slot].values.tolist()))
    tempcard_slot,choose_slot,p_sts,p_card=hit(p_sts,p_card,tempcard_slot,choose_slot,df)
    return tempcard_slot,choose_slot,p_sts,p_card
```
