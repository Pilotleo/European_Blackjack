```python
def card_count(p_card):
    '''
    被呼叫來算點數、大小牌、用掉的牌數、以及有沒有ACE等等
    這裡產出的狀態會放在dataframe裡面，計算包含了BJ、兩張A可拆、一般牌可拆、一般牌可加倍
    paras:
        p_card 傳入要計算的牌組 type:List()，List裡面放card物件
    return
        card_total 總共的牌數 其實都是2
        card_big   總共的大牌
        card_small 總共的小牌
        card_ACE   總共的ACE數
        ACE_Point  依照有沒有ACE動態判斷點數。
    history:2022/10/9 done
    '''
    if p_card[1].point==1 or p_card[0].point==1:
        card_ACE = 1
        ACE_Point=p_card[1].point+p_card[0].point+10
    else:
        card_ACE = 0
        ACE_Point=0
    if p_card[0].weighted ==1:
        card_big = 0
        card_small = 1
    elif p_card[0].weighted == -1:
        card_big = 1
        card_small = 0
    else :
        card_big = 0
        card_small = 0
    if p_card[1].weighted ==1:
        card_big += 0
        card_small += 1
    elif p_card[1].weighted ==-1:
        card_big += 1
        card_small += 0
    else :
        card_big += 0
        card_small += 0    
    card_total=card_big+card_small
    
    return card_total,card_big,card_small,card_ACE,ACE_Point
```

```python
def bf_sts_cal(p_card,bet):
    '''
    用來計算一開始的兩張，總共會帶來的狀態，需要呼叫card_count，去算點數、大小牌、用掉的牌數、以及有沒有ACE等等
    這裡產出的狀態會放在dataframe裡面，計算包含了BJ、兩張A可拆、一般牌可拆、一般牌可加倍
    paras:
        p_card 傳入要計算的牌組 type:List()，List裡面放card物件
        bet    起始的注碼       type:int  
    return
        p_sts  該牌組的狀態     type:dict() 
    history:2022/10/9 done
    '''


    if (p_card[0].point+p_card[1].point==11) and (p_card[0].point==1 or p_card[1].point==1):#只有兩張時加起來十一點且有一張為1，就是BJ
        p_sts={'weight':p_card[0].weighted+p_card[1].weighted,'bet':bet,'card_big':2,'card_small':0,'card_total':2,'card_ACE':1,'card_point':21,'ACE_Point':21,
                  'BJ':'Y','Bust':'N','double_available':'N','Stand':'N','Split_available':'N','Split_count':0,'Split_ACE':'N'
                  ,'surrender':'N','doubled':'N'}
        return p_sts
    elif (p_card[0].point == p_card[1].point) and (p_card[0].point==1 and p_card[1].point==1):#兩張A的拆牌

        
        p_sts={'weight':p_card[0].weighted+p_card[1].weighted,'bet':bet,'card_big':2,'card_small':0,'card_total':2,'card_ACE':2,'card_point':2,'ACE_Point':12,
                  'BJ':'N','Bust':'N','double_available':'N','Stand':'N','Split_available':'Y','Split_count':0,'Split_ACE':'Y'
                  ,'surrender':'N','doubled':'N'}
        return p_sts
    elif (p_card[0].point == p_card[1].point) and (p_card[1].point!=1):#兩張其他的拆牌
        card_total,card_big,card_small,card_ACE,ACE_Point=card_count(p_card)
        p_sts={'weight':p_card[0].weighted+p_card[1].weighted,'bet':bet,'card_big':card_big,'card_small':card_small,
                'card_total':card_total,'card_ACE':card_ACE,'card_point':p_card[1].point*2,'ACE_Point':0,
                  'BJ':'N','Bust':'N','double_available':'N','Stand':'N','Split_available':'Y','Split_count':0,'Split_ACE':'N'
                  ,'surrender':'N','doubled':'N'}
        return p_sts
    else:
        card_total,card_big,card_small,card_ACE,ACE_Point=card_count(p_card)
        p_sts={'weight':p_card[0].weighted+p_card[1].weighted,'bet':bet,'card_big':card_big,'card_small':card_small,
                'card_total':card_total,'card_ACE':card_ACE,'card_point':p_card[1].point+p_card[0].point,'ACE_Point':0,
                  'BJ':'N','Bust':'N','double_available':'Y','Stand':'N','Split_available':'N','Split_count':0,'Split_ACE':'N'
                  ,'surrender':'N','doubled':'N'}
        return p_sts
```

```python
def play_card(p_sts,p_card,d_card,desk,bet):
    ''' ###play_card，主要打牌的行程，
    Paras:
        p_sts, 自已牌的狀態，決定要hit、Stand還是要分牌、double、爆牌等等
        p_card 自己的牌組，傳進來之後也需要靠他再去繼續往下增加牌
        d_card 莊家的牌，會影響自己的打法
    Returns:
        None.
    '''
    if(isinstance(p_card, list) == False):
        return 0
    if p_card ==0 or p_sts['Bust']=='Y' or p_sts['surrender']=='Y'or p_sts['BJ']=='Y':#合條件就不要再跳進來了
        return p_sts

    card_point=0
    card_ACE=0
    ACE_Point=0
    for i in range(len(p_card)):
        if p_card[i].point==1:
            card_ACE += 1
        else:
            card_ACE += 0
        card_point+=p_card[i].point
    if card_ACE>=1:
        ACE_Point=card_point+10
    else:
        ACE_Point=card_point
    p_sts['card_point']=card_point
    p_sts['card_ACE']=card_ACE
    p_sts['ACE_Point']=ACE_Point
    while (p_sts['Bust']=='N'and p_sts['BJ']=='N' and (p_sts['ACE_Point']<=21 or p_sts['card_point']<=21)and p_sts['surrender']=='N' ):
        if p_sts['ACE_Point']>=p_sts['card_point'] and p_sts['card_ACE']>0 and p_sts['ACE_Point']<=21 : # soft_play

            if p_sts['BJ'] == 'Y' or p_sts['Stand'] =='Y':
                prin(p_sts)
                return p_sts
            elif p_sts['ACE_Point']>=19 :
                p_sts['Stand'] ='Y'
                return p_sts
            elif p_sts['ACE_Point']==18 and (d_card[0].point>=9 or d_card[0].point==1):
                p_sts= hit(p_sts,p_card,desk,bet)
            elif p_sts['ACE_Point']==18 and (d_card[0].point==3 or d_card[0].point==4 or d_card[0].point==5 or d_card[0].point==6):
                p_sts=Double(p_sts,p_card,desk,bet,'Stand')
                return p_sts
            elif p_sts['ACE_Point']==18 and (d_card[0].point==8 or d_card[0].point==2 or d_card[0].point==7):
                p_sts['Stand'] ='Y'
                return p_sts
            elif p_sts['ACE_Point']<=17 and (d_card[0].point>=7 or d_card[0].point==1):
                p_sts= hit(p_sts,p_card,desk,bet)
#                 print(p_card)
            elif p_sts['ACE_Point']<=17 and (d_card[0].point==5 or d_card[0].point==6 ):
#                 print(p_card[0],len(p_card),p_sts)
                p_sts=Double(p_sts,p_card,desk,bet,'hit')
                if p_sts['doubled']=='Y':
                    return p_sts
#                 print(p_card[0],len(p_card),p_sts)
            elif p_sts['ACE_Point']>=15 and p_sts['ACE_Point']<=17 and (d_card[0].point==4 or d_card[0].point==5 or d_card[0].point==6 ):
                p_sts=Double(p_sts,p_card,desk,bet,'hit')
                if p_sts['doubled']=='Y':
                    return p_sts
                #print(p_card[0],len(p_card),p_sts)
            elif (p_sts['ACE_Point']==14 or p_sts['ACE_Point']==13) and (d_card[0].point>=4):
                p_sts= hit(p_sts,p_card,desk,bet)
            elif p_sts['ACE_Point']==17 and (d_card[0].point==3):
                p_sts=Double(p_sts,p_card,desk,bet,'hit')
                if p_sts['doubled']=='Y':
                    return p_sts
                #print(p_card[0],len(p_card),p_sts)
            elif p_sts['ACE_Point']<=16 and (d_card[0].point==3):
                p_sts= hit(p_sts,p_card,desk,bet)
            elif p_sts['ACE_Point']<=17 and (d_card[0].point==2):
                p_sts= hit(p_sts,p_card,desk,bet)
            elif p_sts['Split_count']>0:#分過一次的就直接結束
                break
            elif p_sts['ACE_Point']>0:#分過一次的就直接結束
                p_sts= hit(p_sts,p_card,desk,bet)
            else:
                print('Soft Play go wrong','p_sts = ',p_sts,'p_card','d_card=',d_card[0].point)
                for i in range(len(p_card)):
                    print(p_card[i].point)
                break
#                 os._exit()

        else: # Hard_play
#             print('判斷三')
            if p_sts['card_point'] >= 17:
                p_sts['Stand'] ='Y'
                break
            elif p_sts['card_point'] ==16 and (d_card[0].point==9 or d_card[0].point==10):
                p_sts=surrender(p_sts,p_card,desk,bet)
            elif p_sts['card_point'] ==16 and (d_card[0].point==1 or d_card[0].point==8 or d_card[0].point==7):
                p_sts= hit(p_sts,p_card,desk,bet)
            elif p_sts['card_point'] ==16:
                p_sts['Stand'] ='Y'
                break
            elif (p_sts['card_point'] ==15 or p_sts['card_point'] ==14) and (d_card[0].point==10):
                p_sts=surrender(p_sts,p_card,desk,bet)
            elif (p_sts['card_point'] ==15 or p_sts['card_point'] ==14) and (d_card[0].point==1 or d_card[0].point==8 or d_card[0].point==7 or d_card[0].point==9):
                p_sts= hit(p_sts,p_card,desk,bet)
            elif (p_sts['card_point'] ==15 or p_sts['card_point'] ==14):
                p_sts['Stand'] ='Y'
                break
            elif (p_sts['card_point'] ==13 or p_sts['card_point'] ==12)and (d_card[0].point==1 or d_card[0].point==8 or d_card[0].point==7 or d_card[0].point==9 or d_card[0].point==10):
                p_sts= hit(p_sts,p_card,desk,bet)
            elif p_sts['card_point'] ==13:
                p_sts['Stand'] ='Y'
                break
            elif p_sts['card_point'] ==12 and (d_card[0].point==4 or d_card[0].point==5 or d_card[0].point==6):
                p_sts['Stand'] ='Y'
                break
            elif p_sts['card_point'] ==12:
                p_sts= hit(p_sts,p_card,desk,bet)
            elif (p_sts['card_point'] ==10 or p_sts['card_point'] ==11) and(d_card[0].point==1 or d_card[0].point==10):
                p_sts= hit(p_sts,p_card,desk,bet)
            elif (p_sts['card_point'] ==10 or p_sts['card_point'] ==11):
                p_sts=Double(p_sts,p_card,desk,bet,'hit')
                if p_sts['doubled']=='Y':
                    return p_sts
            elif p_sts['card_point'] ==9 and (d_card[0].point==3 or d_card[0].point==4 or d_card[0].point==5 or d_card[0].point==6):
                p_sts=Double(p_sts,p_card,desk,bet,'hit')
                if p_sts['doubled']=='Y':
                    return p_sts
            elif p_sts['card_point'] ==9:
                p_sts= hit(p_sts,p_card,desk,bet)
            elif (p_sts['card_point'] ==4 or p_sts['card_point'] ==6) and (d_card[0].point==8 or d_card[0].point==9 or d_card[0].point==10 or d_card[0].point==1):
                p_sts= hit(p_sts,p_card,desk,bet)
            elif p_sts['card_point'] ==8 and(d_card[0].point==3 or d_card[0].point==4 or d_card[0].point==2 or d_card[0].point==7 or d_card[0].point==1 or d_card[0].point==8 or d_card[0].point==9 or d_card[0].point==10):
                p_sts= hit(p_sts,p_card,desk,bet)
            elif p_sts['card_point'] ==5 or p_sts['card_point'] ==6 or p_sts['card_point'] ==7 or p_sts['card_point'] ==8:
                p_sts= hit(p_sts,p_card,desk,bet)
            elif p_sts['card_point'] ==4:
                p_sts= hit(p_sts,p_card,desk,bet)
            else:

                print('Hard Play go wrong','p_sts = ',p_sts,'d_card=',d_card[0].point)
                for i in range(len(p_card)):
                    print(p_card[i].point)
                break
    return p_sts
    
```

```python
def split(p_sts,p_card,dealer,desk,bet):
    ''' ###split，若依照打法需要分牌，就走分牌，因為要看表，所以不能跟分牌判斷寫在一起。
    Paras:
        p_sts   當下處理牌組的狀況 type:dict
        p_card  玩家手牌     type:list
    Returns:
        None
    '''
    
    if(isinstance(p_card, list) == False):
        return 0
    
    if len(p_card)==2:
        if p_sts['Split_available'] == 'Y' and p_sts['card_ACE']>0 and p_sts['Split_count'] <  1 and p_card[0].point ==p_card[1].point: #AA對拆完補牌後就結束
            if dealer[0].point>=2:
                new1 = desk.pop()
                new2 = desk.pop()
                card_split_org = p_card.pop()
                p_card.append(new1)
                #換牌魔術
                if p_card[0].point!=new1.point:
                    if card_split_org.point==new2.point:
                        card4=p_card.pop()
                        card3=p_card.pop()
                        p_card.append(card_split_org)
                        p_card.append(new2)
                        card_split_org=card3
                        new2=card4

                bf_sts_cal_for_split(p_card,bet)

                p_sts['Split_count']+=1

                if(p_sts['Split_count']>=2):
                    p_sts['Split_available'] = 'N'
                p_sts['Split_ACE']='Y'
                p_sts['doubled']='N'
                return [card_split_org,new2]
            else:
                p_sts['Split_available'] = 'N' #不在這邊處理hit，先讓他離開
                return 0

        elif p_sts['Split_available'] == 'Y' and p_sts['Split_count'] < 2 and p_card[0].point ==p_card[1].point: 

            if (p_card[0].point==10) or (p_card[0].point==9 and (dealer[0].point==7 or dealer[0].point==10 or dealer[0].point==1)):

                p_sts['Split_available']= 'N' 
                p_sts['Stand']='Y'
                p_sts['doubled']='N'
                return 0
            elif ((p_card[0].point==8) or (p_card[0].point==7)) and dealer[0].point==10:

                p_sts['surrender'] = 'Y'
                p_sts['doubled']='N'
                return 0
            elif ((p_card[0].point==9 or
                   (p_card[0].point==8 and dealer[0].point>=2) or     
                  (p_card[0].point==7 and dealer[0].point>=2 and dealer[0].point<=7) or     
                  (p_card[0].point==6 and dealer[0].point>=2 and dealer[0].point<=6) or    
                  (p_card[0].point==4 and dealer[0].point>=5 and dealer[0].point<=6) or    
                  (p_card[0].point==3 and dealer[0].point>=2 and dealer[0].point<=7)or 
                  (p_card[0].point==2 and dealer[0].point>=2 and dealer[0].point<=7))and 
                  p_card[0].point ==p_card[1].point) :

                #把Pair區域裡面的P都塞進來了。

                new1 = desk.pop()
                new2 = desk.pop()
                card_split_org = p_card.pop()
                p_card.append(new1)
                
                #換牌魔術
                if p_card[0].point!=new1.point:
                    if card_split_org.point==new2.point:
                        card4=p_card.pop()
                        card3=p_card.pop()
                        p_card.append(card_split_org)
                        p_card.append(new2)
                        card_split_org=card3
                        new2=card4
                
                bf_sts_cal_for_split(p_card,bet)
                
                if p_card[0].point!=card_split_org.point:
                    print('分之前的牌1',p_card[0].point,'新牌1',new1.point,'分之前的牌1',card_split_org.point,'新牌2', new2.point)
                p_sts['Split_count']+=1
                p_sts['doubled']='N'

                if(p_sts['Split_count']>=2):
                    p_sts['Split_available'] = 'N'
                return [card_split_org,new2]
        else:
            return 0
    else:
        return 0
```

```python


def hit_card_count(p_card):
    '''
    被呼叫來算點數、大小牌、用掉的牌數、以及有沒有ACE等等
    這裡產出的狀態會放在dataframe裡面，計算包含了BJ、兩張A可拆、一般牌可拆、一般牌可加倍
    paras:
        p_card 傳入要計算的牌組 type:List()，List裡面放card物件
    return
        card_total 總共的牌數 其實都是2
        card_big   總共的大牌
        card_small 總共的小牌
        card_ACE   總共的ACE數
        ACE_Point  依照有沒有ACE動態判斷點數。
    history:2022/10/9 done
    '''
    card_ACE= 0
    card_big=0
    card_small=0
    total_weight=0
    card_point=0
    for i in range(len(p_card)) :
        if p_card[i].point==1:
            card_ACE+=1
    if card_ACE>0:
        for i in range(len(p_card)) :
            card_point+=p_card[i].point
        ACE_Point=card_point+10
    else:
        for i in range(len(p_card)) :
            card_point+=p_card[i].point
        ACE_Point=card_point
    for i in range(len(p_card)) :
        if p_card[i].weighted ==1:
            card_big += 0
            card_small += 1
        elif p_card[i].weighted ==-1:
            card_big += 1
            card_small += 0
        else :
            card_big += 0
            card_small += 0    
        total_weight+=p_card[i].weighted
    card_total=len(p_card)
    
    return card_total,card_big,card_small,card_ACE,ACE_Point,card_point,total_weight
```

```python
def hit_sts_cal(p_sts,p_card,bet):
    '''
    用來計算加牌後的計算，總共會帶來的狀態，需要呼叫card_count，去算點數、大小牌、用掉的牌數、以及有沒有ACE等等
    這裡產出的狀態會放在dataframe裡面，計算包含了BJ、兩張A可拆、一般牌可拆、一般牌可加倍
    !!!分牌後沒有BJ，此處也與原本的狀態計算不同
    paras:
        p_card 傳入要計算的牌組 type:List()，List裡面放card物件
        bet    起始的注碼       type:int  
    return
        p_sts  該牌組的狀態     type:dict() 
    history:2022/10/9 done
    '''
    card_total,card_big,card_small,card_ACE,ACE_Point,card_point,total_weight=hit_card_count(p_card)
    
    if card_point>21:
        p_sts={'weight':total_weight,'bet':bet,'card_big':card_big,'card_small':card_small,'card_total':card_total,
               'card_ACE':card_ACE,'card_point':card_point,'ACE_Point':ACE_Point,
                  'BJ':'N','Bust':'Y','double_available':'N','Stand':'N','Split_available':'N','Split_count':p_sts['Split_count'],'Split_ACE':p_sts['Split_ACE']
                  ,'surrender':'N','doubled':p_sts['doubled']}
    elif card_point<=21 and ACE_Point>21:
        p_sts={'weight':total_weight,'bet':bet,'card_big':card_big,'card_small':card_small,'card_total':card_total,
               'card_ACE':card_ACE,'card_point':card_point,'ACE_Point':ACE_Point,
                  'BJ':'N','Bust':'N','double_available':'N','Stand':'N','Split_available':'N','Split_count':p_sts['Split_count'],'Split_ACE':p_sts['Split_ACE']
                  ,'surrender':'N','doubled':p_sts['doubled']}
    else:#ACE_Point=card_point
        p_sts={'weight':total_weight,'bet':bet,'card_big':card_big,'card_small':card_small,'card_total':card_total,
               'card_ACE':card_ACE,'card_point':card_point,'ACE_Point':ACE_Point,
                  'BJ':'N','Bust':'N','double_available':'N','Stand':'N','Split_available':'N','Split_count':p_sts['Split_count'],'Split_ACE':p_sts['Split_ACE']
                  ,'surrender':'N','doubled':p_sts['doubled']}
    return p_sts
```

```python
def after_play_cal(p_card,p_sts):
    if(isinstance(p_card, list) == False):
        return 0
    card_point=0
    card_ACE=0
    ACE_Point=0
    weight=0
    card_big=0
    card_small=0
    for i in range(len(p_card)):
        if p_card[i].point==1:
            card_ACE += 1
        else:
            card_ACE += 0
        if p_card[i].weighted==1:
            card_small+=1
        elif p_card[i].weighted==-1:
            card_big+=1
        card_point+=p_card[i].point
        weight+=p_card[i].weighted
    if card_ACE>=1:
        ACE_Point=card_point+10
    else:
        ACE_Point=card_point
    card_total = len(p_card)
    
    if (card_point>21):#只有兩張時加起來十一點且有一張為1，就是BJ
        p_sts={'weight':weight,'bet':0,'card_big':card_big,'card_small':card_small,'card_total':card_total,'card_ACE':card_ACE,'card_point':card_point,'ACE_Point':ACE_Point,
                  'BJ':'N','Bust':'Y','double_available':'N','Stand':p_sts['Stand'],'Split_available':'N','Split_count':p_sts['Split_count'],'Split_ACE':p_sts['Split_ACE']
                  ,'surrender':'N','doubled':p_sts['doubled']}
        return p_sts
    elif (ACE_Point==21 and len(p_card)==2) and p_sts['Split_count']==0:#只有兩張時加起來十一點且有一張為1，就是BJ
        p_sts={'weight':weight,'bet':0,'card_big':card_big,'card_small':card_small,'card_total':card_total,'card_ACE':card_ACE,'card_point':card_point,'ACE_Point':ACE_Point,
                  'BJ':'Y','Bust':'N','double_available':'N','Stand':p_sts['Stand'],'Split_available':'N','Split_count':p_sts['Split_count'],'Split_ACE':p_sts['Split_ACE']
                  ,'surrender':'N','doubled':p_sts['doubled']}
        return p_sts
    else:
        p_sts={'weight':weight,'bet':0,'card_big':card_big,'card_small':card_small,'card_total':card_total,'card_ACE':card_ACE,'card_point':card_point,'ACE_Point':ACE_Point,
                  'BJ':'N','Bust':'N','double_available':'N','Stand':'Y','Split_available':'N','Split_count':p_sts['Split_count'],'Split_ACE':p_sts['Split_ACE']
                  ,'surrender':p_sts['surrender'],'doubled':p_sts['doubled']}
        return p_sts
```

```python
def close_to_zero(row):
    
    if(row==0):
        return row
    elif(row<0):
        return row+1
    else:
        return row-1
    
def true_num_cal(weight):
#     if int(weight/6)>=6:
#         true_num =6
#     elif int(weight/6)==5:
#         true_num =5
    if int(weight/6)>=4:
        true_num =4
    elif int(weight/6)==3:
        true_num =3
    elif int(weight/6)==2:
        true_num =2
    elif int(weight/6)==1:
        true_num =1
    else:
        true_num =0
    return true_num
```

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def calculate_pval(row):
    nobs = np.array([row['Total Card'], row['Total Card']])
    count = np.array([row['Small Card'], row['Big Card']])
    stat, pval = proportions_ztest(count, nobs)
    return pval

true_num_df1['Weighted_mean'] = true_num_df1['Mean Profit']*true_num_df1['總場數']
true_num_df1['Weighted_ROC'] = true_num_df1['ROC']*true_num_df1['總場數']

res = true_num_df1.groupby('True Number').agg({'Weighted_mean':'sum',
                                        '勝場':'sum',
                                        '敗場':'sum',
                                         '總場數':'sum',
                                        'Small Card':'sum',
                                        'Big Card':'sum',
                                        'Total Card':'sum',
                                        'Weighted_ROC':'sum'}).reset_index()

res['平均利潤'] = res['Weighted_mean']/res['總場數']
res['投資報酬率'] = res['Weighted_ROC']/res['總場數']

## calculate SD

tmp_for_calculation = true_num_df1.merge(res[['True Number','平均利潤']], on ='True Number')
tmp_for_calculation['SD_Tmp'] = ((tmp_for_calculation['Mean Profit'] - tmp_for_calculation['平均利潤'])**2 + tmp_for_calculation['STD']**2)* tmp_for_calculation['總場數']


tmp_for_calculation = tmp_for_calculation.groupby('True Number').agg({'SD_Tmp':'sum'}).reset_index()

res = res.merge(tmp_for_calculation, on ='True Number')

res['SD'] = (res['SD_Tmp']/ res['總場數']) ** (1/2)


res = res.drop(columns = ['Weighted_mean','SD_Tmp'])

res['大牌比率'] = res['Big Card']/res['Total Card']

res['小牌比率'] = res['Small Card']/res['Total Card']


res['pvalue'] = res.apply(lambda row:calculate_pval(row),axis = 1)
```
